```python
from bioservices import KEGG
```

First, we create KEGG object. KEGG object has many useful functions and attributes that we can use to access KEGG database.

```python
k = KEGG()
```

List all organisms in KEGG database.

```python
print k.list('organism')
```

List all human pathways.

```python
print k.list('pathway', organism='hsa')
```

The total number of human pathways.

```python
hsa_pathways = [x for x in k.list('pathway', organism='hsa').split('\n')]
print len(hsa_pathways)
```

List 5 organism IDs.

```python
k.organismIds[:5]
```

Search for an organism using a keyword.

```python
k.lookfor_organism('droso')
```

Search for a pathway using a keyword.

```python
k.lookfor_pathway('apoptosis')
```

Or

```python
print k.find('pathway', 'apoptosis')
#add find parameters list
```

Look for "MYC" gene in human.

```python
print k.find('hsa', 'MYC')
```

Look for pathways containing "MYC".

```python
k.get_pathway_by_gene('4609', 'hsa') # use gene ID instead of gene name
                                     # either gene name or gene ID can be used
```

Get info about a specific pathway.

```python
print k.get('hsa04010') # use pathway Id
```

Parse data to dictionary using KEGGparser

```python
from bioservices import KEGGParser
```

```python
parser = KEGGParser()
```

```python
data = k.get('hsa04010')
dict_data = parser.parse(data)
```

```python
dict_data.keys()
```

```python
print dict_data['NAME']
for key in dict_data['GENE']:
    print key, dict_data['GENE'][key]
```

Download and parse data in KGML format.

```python
res = k.parse_kgml_pathway('hsa04010')
```

Inspect the first relationship.

```python
res['relations'][3]
```

Inspect the first entry.

```python
res['entries'][38]
```
